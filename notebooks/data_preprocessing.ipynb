```python
import os
import json
```

```python
# Funktion zum Korrigieren des Encodings
def correct_encoding(data):
    if isinstance(data, str):
        try:
            # Versuche, die Zeichenkette zuerst zu dekodieren, dann wieder zu encodieren und schließlich erneut zu dekodieren
            return data.encode('utf-8').decode('latin1').encode('latin1').decode('utf-8')
        except UnicodeEncodeError:
            return data  # Falls die Zeichenkette nicht verarbeitet werden kann, zurückgeben
        except UnicodeDecodeError:
            return data  # Falls die Zeichenkette nicht verarbeitet werden kann, zurückgeben
    elif isinstance(data, list):
        return [correct_encoding(item) for item in data]
    elif isinstance(data, dict):
        return {key: correct_encoding(value) for key, value in data.items()}
    return data
```

```python
# Lade die gescrapten Daten
with open('../data/raw/scraped_data.json', 'r', encoding='utf-8') as file:
    raw_data = json.load(file)
```

```python
processed_data = correct_encoding(raw_data)
```

```python
# Speichere die korrekt dekodierten Daten
with open('../data/processed/processed_data.json', 'w', encoding='utf-8') as file:
    json.dump(processed_data, file, indent=4, ensure_ascii=False)
```

```python
# Beispielausgabe der korrekt dekodierten Daten
for item in processed_data:
    print(f"Title: {item['title']}")
    print(f"Content: {item['content']}\n")
```
